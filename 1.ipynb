````python
import markdown2
import os

# Basic Markdown to HTML conversion
def basic_markdown_conversion():
    """
    Demonstrates basic Markdown to HTML conversion
    """
    markdown_text = """
# Welcome to Markdown

## Subheading

This is a **bold** and *italic* text example.

- List item 1
- List item 2

[Link to Example](https://example.com)

\`\`\`python
def hello_world():
    print("Hello, Markdown!")
\`\`\`
"""
    
    # Convert Markdown to HTML
    html_content = markdown2.markdown(markdown_text)
    print(html_content)

# Advanced Markdown conversion with extras
def advanced_markdown_conversion():
    """
    Demonstrates advanced Markdown conversion with extra features
    """
    markdown_text = """
# Advanced Markdown

## Code Syntax Highlighting

Python code:
\`\`\`python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
\`\`\`

## Task Lists

- [x] Completed task
- [ ] Pending task
"""
    
    # Convert with extras like syntax highlighting, task lists
    html_content = markdown2.markdown(
        markdown_text, 
        extras=[
            "code-friendly",     # Better code block handling
            "fenced-code-blocks",# Support for ```language code blocks
            "cuddled-lists",     # More flexible list parsing
            "task-lists"         # GitHub-style task lists
        ]
    )
    print(html_content)

# Markdown to HTML file conversion
def markdown_to_html_file(markdown_file, output_file):
    """
    Convert a Markdown file to an HTML file
    
    Args:
        markdown_file (str): Path to input Markdown file
        output_file (str): Path to output HTML file
    """
    # Read Markdown file
    with open(markdown_file, 'r', encoding='utf-8') as f:
        markdown_content = f.read()
    
    # Convert to HTML
    html_content = markdown2.markdown(
        markdown_content, 
        extras=["metadata", "code-friendly"]
    )
    
    # Write to HTML file
    with open(output_file, 'w', encoding='utf-8') as f:
        # Optional: Add basic HTML structure
        full_html = f"""
        <!DOCTYPE html>
        <html>
        <head>
            <title>Markdown Converted Document</title>
            <style>
                body {{ font-family: Arial, sans-serif; line-height: 1.6; max-width: 800px; margin: 0 auto; }}
                code {{ background-color: #f4f4f4; padding: 2px 4px; }}
                pre {{ background-color: #f1f1f1; padding: 10px; }}
            </style>
        </head>
        <body>
            {html_content}
        </body>
        </html>
        """
        f.write(full_html)
    
    print(f"Converted {markdown_file} to {output_file}")

# Example usage
if __name__ == "__main__":
    # Uncomment to test different functions
    # basic_markdown_conversion()
    # advanced_markdown_conversion()
    
    # Example of converting a file
    # Ensure you have a markdown file named 'example.md' in the same directory
    # markdown_to_html_file('example.md', 'output.html')
    pass

# Installation instructions
"""
To use this script:

1. Install markdown2:
   pip install markdown2

2. Run the script and uncomment the function calls you want to test
"""
````
